```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import (
    train_test_split, KFold, cross_val_score, RandomizedSearchCV, GridSearchCV
)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, precision_recall_curve,
    roc_curve, auc
)

%matplotlib inline
```

```python
data = pd.read_csv("indian_liver_patient.csv")
```

```python
data.info()
```

```python
missing = data[data.Albumin_and_Globulin_Ratio.isna()]

for index, row in missing.iterrows():
    age, disease, gender = row["Age"], row["Dataset"], row["Gender"]
    new_table = data[(data["Age"] == age) & (data["Gender"] == gender) & (data["Dataset"] == disease)]
    print(age, disease, gender, new_table["Albumin_and_Globulin_Ratio"].mean())
    data.at[index, "Albumin_and_Globulin_Ratio"] = new_table["Albumin_and_Globulin_Ratio"].mean()
```

```python
data = data.rename(columns = {"Alkaline_Phosphotase": "ALP", 
                              "Alamine_Aminotransferase": "ALT", 
                              "Aspartate_Aminotransferase": "AST", 
                              "Total_Protiens":"Protein", "Albumin_and_Globulin_Ratio": "AGR", 
                              "Dataset": "Liver Patient"})

data["Liver Patient"].replace(2, 0, inplace = True)
```

```python
data.info()
```

```python
data = pd.get_dummies(data, columns = ["Gender"], drop_first = True)
```

```python
data.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.drop(["Liver Patient"], axis = 1),
                                                    data["Liver Patient"],
                                                    test_size = 0.3,
                                                    random_state = 1)
```

# Logistic Regression without Hyperparameter Tuning

```python
lr = LogisticRegression(solver = 'liblinear', C = 1.0, max_iter = 1000)
lr.fit(x_train, y_train)
y_prob_train = lr.predict_proba(x_test)[:, 1]
```

```python
fpr, tpr, thresh_roc = roc_curve(y_test, y_prob_train)
prec, rec, thresh_pr = precision_recall_curve(y_test, y_prob_train)
```

```python
sns.set_style('darkgrid')
```

```python
no_skill = len(y_test[y_test == 1]) / len(y_test)

plt.figure(figsize = (12, 8))
plt.plot([0, 1], [no_skill, no_skill], linestyle = '--', label = 'Random Chance', color = 'blue')
plt.plot(rec, prec, marker = '.', label='Logistic', color = 'orange')

plt.xlabel('Recall')
plt.ylabel('Precision')

plt.legend()

plt.show()
```

```python
auc(rec, prec)
```

```python
probs_ns = [0 for _ in range(len(y_test))]

fpr_ns, tpr_ns, thresh_ns = roc_curve(y_test, probs_ns)

plt.figure(figsize = (12, 8))
plt.plot(fpr_ns, tpr_ns, linestyle = '--', label = 'Random Chance', color = 'blue')
plt.plot(fpr, tpr, marker = '.', label = 'Logistic', color = 'orange')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend()

plt.show()
```

```python
roc_auc_score(y_test, y_prob_train)
```

```python
report = classification_report(y_test, lr.predict(x_test), output_dict = True)
report_simple = {"No Liver Disease": report["0"], "Liver Disease": report["1"]}
pd.DataFrame(report_simple)
```

The Precision-Recall curve and ROC suggest reasonable performance for the logistic regression model without any hyperparameter tuning. The classification report reveals that much of the model's weakness comes from identifying No Liver Disease which is fair because there's a lot fewer observations of No Liver Disease in the dataset.

# Model Shootout - RFC and Logistic Regression

```python
def compare_models(models, X_data, y_data, n_fold, score):
    for name, model in models.items():
        kfold = KFold(n_splits = n_fold)
        cv_results = cross_val_score(model, X_data, y_data, cv = kfold, scoring = score)
        eval_msg = f"{name}: {cv_results.mean()} ({cv_results.std()})"
        print(eval_msg)
```

```python
compare_models({'rfc':RandomForestClassifier(), 'lr':LogisticRegression(max_iter = 1000)}, 
               x_train, y_train, 10, 'roc_auc')
```

Because the standard deviations cause the confidence intervals to overlap, the models are more or less equivalent. Nonetheless, I'll try tuning the hyperparameters on both models to see if any differences can be found there.

# Hyperparameter Tuning - RFC

```python
def random_hyper_RFC(X_Train, Y_Train, cross_val = 5):
    hyperparameters = {'n_estimators': [100, 300, 500, 800, 1200], 
                       'max_depth': [5, 10, 15, 25, 30],
                       'min_samples_split': [2, 5, 10, 25, 100],
                       'min_samples_leaf': [1, 2, 5, 10]}
    random_search = RandomizedSearchCV(RandomForestClassifier(), hyperparameters, cv = cross_val)
    best_fit = random_search.fit(X_Train, Y_Train)
    
    return best_fit.best_params_
```

```python
new_RFC_params = random_hyper_RFC(x_train, y_train)
tuned_RFC = RandomForestClassifier(**new_RFC_params)
tuned_RFC.fit(x_train, y_train)
```

```python
tuned_rfc_prob = tuned_RFC.predict_proba(x_test)[:, 1]
fpr3, tpr3, thresh_roc3 = roc_curve(y_test, tuned_rfc_prob)
prec3, rec3, thresh_pr3 = precision_recall_curve(y_test, tuned_rfc_prob)
```

```python
plt.figure(figsize = (12, 8))
plt.plot([0, 1], [no_skill, no_skill], linestyle = '--', label = 'Random Chance', color = 'blue')
plt.plot(rec3, prec3, marker = '.', label='RFC', color = 'orange')

plt.xlabel('Recall')
plt.ylabel('Precision')

plt.legend()

plt.show()
```

```python
auc(rec3, prec3)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(fpr_ns, tpr_ns, linestyle = '--', label = 'Random Chance', color = 'blue')
plt.plot(fpr3, tpr3, marker = '.', label = 'RFC', color = 'orange')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend()

plt.show()
```

```python
roc_auc_score(y_test, tuned_rfc_prob)
```

```python
report3 = classification_report(y_test, tuned_RFC.predict(x_test), output_dict = True)
report_simple3 = {"No Liver Disease": report3["0"], "Liver Disease": report3["1"]}
pd.DataFrame(report_simple3)
```

The tuned RFC does improve a bit over the default model, most interestingly in the no liver disease category. Precision is much higher than the default model for RFC in identifying no liver disease but comes at the great expense of recall. That is to say, of the observations flagged as having no liver disease, a great proportion of them truly do not have liver disease but this comes at the expense of incorrectly missing more people who don't have liver disease.

The recall of liver disease is greatly improved vs the default model as well.
